```python
import re
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.corpus import stopwords
```

## Load Data

```python
data = pd.read_csv("data/tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

## Lowercase

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data['review_lowercase']
```

```python
data.head()
```

## Stop word removal

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove('not')
```

```python
data['review_no_stopwords'] = data['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
```

```python
data['review_no_stopwords'][0]
```

## Punctuation
We want to remove punctation but we notice we have a few reviews with the symbol * instead of the word sta so we replace it with word form.

```python
data['review_no_stopwords_no_punct'] = data.apply(lambda x: re.sub(r"[*]", "star", x["review_no_stopwords"]), axis=1)
```

```python
data.head()
```

```python
# remove everything but words and spaces (punctuation)
data['review_no_stopwords_no_punct'] = data.apply(lambda x: re.sub(r"[^\w\s]", "", x["review_no_stopwords"]), axis=1)
```

```python
data.head()
```

## Tokenizing

```python
data['tokenized'] = data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punct']), axis=1)
```

```python
data.tokenized[0]
```

## Stemming

```python
ps = PorterStemmer()
```

```python
data["stemmed"] = data["tokenized"].apply(lambda tokens: [ps.stem(t) for t in tokens])
```

```python
data["stemmed"][0]
```

## Lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
data["lemmatized"] = data["tokenized"].apply(lambda tokens: [lemmatizer.lemmatize(t) for t in tokens])
```

```python
data["lemmatized"][0]
```

## N-grams

```python
import itertools
tokens_clean = list(itertools.chain.from_iterable(data['lemmatized']))

# tokens_clean = sum(data['lemmatized'], [])
```

```python
len(tokens_clean), tokens_clean
```

```python
# unigrams: n=1
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
```

```python
print(unigrams)
```

```python
# bigrams: n=2
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
```

```python
print(bigrams)
```

```python
# trigrams: n=3
trigrams = (pd.Series(nltk.ngrams(tokens_clean, 3)).value_counts())
```

```python
print(trigrams)
```
