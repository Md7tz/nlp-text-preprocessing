```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tokens = [
    'the', 'rise', 'of', 'artificial', 'intelligence', 'and', 'machine', 'learning', 
    'has', 'transformed', 'multiple', 'industries', 'by', 'automating', 'tasks', 
    'enhancing', 'data', 'analysis', 'and', 'improving', 'decision', 'making', 
    'capabilities', '.', 'natural', 'language', 'processing', 'is', 'one', 'of', 
    'the', 'key', 'fields', 'within', 'AI', 'enabling', 'computers', 'to', 
    'understand', 'interpret', 'and', 'generate', 'human', 'language', '.', 
    'applications', 'such', 'as', 'chatbots', 'language', 'translation', 'sentiment', 
    'analysis', 'and', 'text', 'summarization', 'are', 'now', 'commonplace', 
    'in', 'both', 'consumer', 'and', 'enterprise', 'environments', '.', 'with', 
    'the', 'availability', 'of', 'large', 'datasets', 'and', 'powerful', 
    'computational', 'resources', ',', 'the', 'development', 'of', 'deep', 
    'learning', 'models', 'continues', 'to', 'accelerate', ',', 'paving', 'the', 
    'way', 'for', 'more', 'intelligent', 'and', 'context-aware', 'systems', '.'
]
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens, 1)).value_counts())
```

```python
print(unigrams[:10])
```

```python
unigrams[:10].sort_values().plot.barh(color="green", width=.8, figsize=(12,4))
plt.title("10 Most Frequently Occuring Unigrams")
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())
print(bigrams[:10])
```

```python
trigrams = (pd.Series(nltk.ngrams(tokens, 3)).value_counts())
print(trigrams[:10])
```
